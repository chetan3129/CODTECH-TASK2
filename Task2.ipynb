```python
import ipywidgets as widgets
from IPython.display import display

print('BMI Calculator')
height_widget = widgets.FloatText(description="Height (m):")
weight_widget = widgets.FloatText(description="Weight (kg):")
calculate_button = widgets.Button(description="Calculate")
output_widget = widgets.Label()

def calculate_bmi(button):
  height = height_widget.value
  weight = weight_widget.value
  if height > 0 and weight > 0:
    bmi = weight / (height ** 2)
    output_widget.value = f"Your BMI is: {bmi:.2f}"
  else:
    output_widget.value = "Please enter valid values."

calculate_button.on_click(calculate_bmi)
display(height_widget, weight_widget, calculate_button, output_widget)
```
